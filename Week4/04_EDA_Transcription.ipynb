📌한번 같이 생각해봐요!📌  
아래는 Age Column에 비어있는 177개의 값을 채우기 위한 아이디어 입니다.  
  
이 분은 Age에 비어 있는 값을 채워주기 위해, 평균값을 이용하여 채워주려고 했지만,
위의 시각화 자료를 보아하니, 연령대의 분포가 너무 다양하여 평균값 하나로만 채우기 애매했죠.
단순히 평균값으로 채워버리면, 4세의 정보에 평균값인 29살을 넣어버리는 경우가 발생할 수 있었습니다.
  
이를 해결하기 위한 아이디어로 저자는 `Name`에서 그 힌트를 활용하기로 합니다.  
`Name`을 잘 보면, *mr*, *mrs* 와 같이 이름에 대한 정보로 데이터를 구분할 수 있다는 사실을 알아낼 수 있습니다.  
  
즉!!!! Age의 null 값을 단순 평균치가 아닌, 이름에서 힌트를 얻어 이들의 평균값으로 대체하면, 조금 더 세밀한 결측치 제거가 가능하다는 것이죠!  
  
한번 어떻게 했는지 따라가봅시다 :)


````python
## 먼저, 이름에서 특정한 부분을 추출해봅시다.
data['Initial']=0
for i in data:
    data['Initial']=data.Name.str.extract('([A-Za-z]+)\.') # 이 정규식은 무엇을 의미할까요? 헷갈리면... 정규식 고수 유우비트님께 물어봅시다..! ㅎㅎ
data
````

````python
data['Initial'] = 0
for i in data :
    data['Initial'] = data.Name.str.extract('([A-Za-z]+)\.')
data
````

````python
## 뽑아낸 이니셜이 성별로 어떤 분포를 가지고 있는지 표로 살펴봅시다
pd.crosstab(data.Initial,data.Sex).T 
````

````python
pd.crosstab(data.Initial, data.Sex).T
````

````python
## 일부 발생한 오타를 replace 함수를 통해 정리해줍니다.
data['Initial'].replace(
    ['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],
    ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],
    inplace=True)
print(data.groupby('Initial')['Age'].mean()) # 평균값 확인!
````

````python
data['Initial'].replace(
    ['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],
    ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],
    inplace=True
)
print(data.groupby('Initial')['Age'].mean())
````

````python
# 이제 각 이니셜 별로 평균 나이를 알아냈으니, 각 평균값에 맞게 채워줍니다.
# pandas에서 배웠던 loc는 이런 식으로 활용됩니다 :)
data.loc[(data.Age.isnull())&(data.Initial=='Mr'),'Age']=33
data.loc[(data.Age.isnull())&(data.Initial=='Mrs'),'Age']=36
data.loc[(data.Age.isnull())&(data.Initial=='Master'),'Age']=5
data.loc[(data.Age.isnull())&(data.Initial=='Miss'),'Age']=22
data.loc[(data.Age.isnull())&(data.Initial=='Other'),'Age']=46

print("결측치 제거 후 남은 결측치: ",data.Age.isnull().sum())
````

````python
data.loc[(data.Age.isnull()) & (data.Initial=='Mr'), 'Age']=33
data.loc[(data.Age.isnull()) & (data.Initial=='Mrs'), 'Age']=36
data.loc[(data.Age.isnull()) & (data.Initial=='Master'), 'Age']=5
data.loc[(data.Age.isnull()) & (data.Initial=='Miss'), 'Age']=22
data.loc[(data.Age.isnull()) & (data.Initial=='Other'), 'Age']=46

print('결측치 제거 후 남은 결측치: ', data.Age.isnull().sum())
````
